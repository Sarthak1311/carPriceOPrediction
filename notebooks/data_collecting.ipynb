```python
def extract_data(url):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text
    soup = BeautifulSoup(webpage)
    # init the lists 
    Dom ,company,model, odometer_reading,fuel_type,transmission,RTo= [],[],[],[],[],[],[]
    
    # getting the name model and manufacture of the car 
    for i in soup.find_all('span',class_ ='sc-braxZu kjFjan'):
        text = i.text.split(' ')
        Dom.append(text[0])
        company.append(text[1])
        model.append(" ".join(text[2:]))

    # loading the container info of page 
    car_info = soup.find_all('div',class_='styles_outer__NTVth')
    data =[]

    # loading all the content text in data 
    for i in car_info:
        for tag in i.find_all('p', class_="sc-braxZu kvfdZL"):
            data.append(tag.text)

    # seprating the data into dometer_reading,fuel_type,transmission,RTo 

    for idx , item in enumerate(data,start=1):
        if idx % 4 ==1:
            odometer_reading.append(item)
        elif idx % 4 ==2:
            fuel_type.append(item)
        elif idx % 4 ==3:
            transmission.append(item)
        else:
            RTo.append(item)

    
    return Dom ,company,model, odometer_reading,fuel_type,transmission,RTo

        
    

    
    
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
```

```python
def extract_data(url):
    # Setup headless Chrome
    options = Options()
    options.add_argument("--headless")  # comment out this line if you want to see the browser
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    driver.get(url)

    # Scroll to load all content
    for _ in range(25):  # increase this number if you still don't get all cars
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # wait for content to load

    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()

    # Init lists
    Dom, company, model = [], [], []
    odometer_reading, fuel_type, transmission, RTo = [], [], [], []

    # Get name, model, manufacture
    for i in soup.find_all('span', class_='sc-braxZu kjFjan'):
        text = i.text.split(' ')
        if len(text) >= 3:
            Dom.append(text[0])
            company.append(text[1])
            model.append(" ".join(text[2:]))
        else:
            # Handle broken titles
            Dom.append("")
            company.append("")
            model.append(i.text)

    # Get detailed car info
    car_info = soup.find_all('div', class_='styles_outer__NTVth')
    data = []

    for i in car_info:
        for tag in i.find_all('p', class_="sc-braxZu kvfdZL"):
            data.append(tag.text)

    # Separate data into odometer, fuel, transmission, RTO
    for idx, item in enumerate(data, start=1):
        if idx % 4 == 1:
            odometer_reading.append(item)
        elif idx % 4 == 2:
            fuel_type.append(item)
        elif idx % 4 == 3:
            transmission.append(item)
        else:
            RTo.append(item)

    return Dom, company, model, odometer_reading, fuel_type, transmission, RTo
```

```python
url = "https://www.cars24.com/buy-used-petrol-cars-new-delhi/?sort=bestmatch&serveWarrantyCount=true&storeCityId=2"
Dom, company, model, odo, fuel, trans, rto = extract_data(url)

print("Total cars scraped:", len(model))
```
